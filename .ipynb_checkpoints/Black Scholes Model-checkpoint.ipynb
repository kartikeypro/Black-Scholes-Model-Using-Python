```python
import numpy as np

import scipy.stats as si

from sympy.stats import Normal, cdf

def euro_call(S0, K, T, r, sigma):

    #S0: Price at time = 0 of the underlying asset - Note that St is log-normally distributed
    #K: Strike price
    #T: Time to option maturity, in units of a year
    #r: Continuously-compounded annual risk-less interest rate
    #sigma: Volatility of annual returns on the underlying asset

    x = np.log(S0 / (K * np.exp(-r*T))) / (sigma * np.sqrt(T)) + 0.5 * sigma * np.sqrt(T)

    call = (S0 * si.norm.cdf(x, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(x - sigma * np.sqrt(T), 0.0, 1.0))

    return call

for spot_price in range(200):

    print (spot_price, euro_call(spot_price, 75, 2, 0.1, 0.25))
```
